# Image To pencil Sketch in Python

## Convert Image to Pencil Sketch

## Convert to Grey Image

```python
grey_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(grey_img)
plt.axis(False)
plt.show()
```

## Invert Image

```python
invert_img = cv2.bitwise_not(grey_img)
```

```python
plt.imshow(invert_img)
plt.axis(False)
plt.show()
```

## Blur Image

```python
blur_img = cv2.GaussianBlur(invert_img, (111,111),0)
```

```python
plt.imshow(blur_img)
plt.axis(False)
plt.show()
```

## Invert Blur Image

```python
invert_blur_img = cv2.bitwise_not(blur_img)
```

```python
plt.imshow(invert_blur_img)
plt.axis(False)
plt.show()
```

## Sketch

```python
sketch_img = cv2.divide(grey_img, invert_blur_img, scale= 256.0) 
```

## Display Sketch

```python
cv2.imshow("sketch image",sketch_img)
cv2.waitKey(0)
cv2.destroyAllWindows()    #This will open a new window title as Sketch image
```

## Save Sketch

```python
 cv2.imwrite("sketch.png", sketch_img)
```

## Create Function 

```python
def img2sketch(photo, k_size):
    #Read Image
    img=cv2.imread(photo)
    
    # Convert to Grey Image
    grey_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Invert Image
    invert_img=cv2.bitwise_not(grey_img)
    #invert_img=255-grey_img

    # Blur image
    blur_img=cv2.GaussianBlur(invert_img, (k_size,k_size),0)

    # Invert Blurred Image
    invblur_img=cv2.bitwise_not(blur_img)
    #invblur_img=255-blur_img

    # Sketch Image
    sketch_img=cv2.divide(grey_img,invblur_img, scale=256.0)

    # Save Sketch 
    cv2.imwrite('sketch.png', sketch_img)

    # Display sketch
    cv2.imshow('sketch image',sketch_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
#Function call
img2sketch(photo='Ranbir.jpg', k_size=7)
```
